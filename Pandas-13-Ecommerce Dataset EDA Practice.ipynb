```python
# find out stocks which are more than 10 quantity and are from United Kingdom
df[(df['Quantity'] >= 10) & (df['Country'] == 'United Kingdom')]
```

```python
# Total Sales per Invoice 
df['Total_Sales'] = df['Quantity'] * df['UnitPrice']
df.groupby('InvoiceNo')['Total_Sales'].sum().reset_index()
```

```python
df.head(2)
```

```python
#Find the top 5 products by total quantity sold.
df.groupby('Description')['Quantity'].sum().reset_index().sort_values(by = 'Quantity', ascending = False).head()
```

```python
# Total Revenue by Country - Aggregate total sales by Country and sort them in descending order.
df.groupby('Country')['Total_Sales'].sum().sort_values(ascending = False).reset_index()
```

```python
# Average Basket Value - Calculate the average amount per invoice. - total revenue divided by number of invoices

total_sum = df['Total_Sales'].sum()
total_invoice = df['InvoiceNo'].count()

print('average amount per invoice is ', total_sum/total_invoice)
```

```python
# For each product (StockCode), compute: Total quantity sold, Total revenue, Average unit price

df.groupby(['StockCode','Description']).agg(Total_Quantity_Sold = ('Quantity', 'sum'),
                              Total_Revenue = ('Total_Sales', 'sum'),
                               Average_Unit_Price = ('UnitPrice', 'mean')
)
```

```python
# Most Popular Product per Country (by revenue)
df.groupby(['Country', 'Description'], as_index=False).agg(Total_Sales=('Total_Sales', 'sum')).sort_values(['Country', 'Total_Sales'], ascending=[True, False]).drop_duplicates(subset = ['Country'], keep='first')
```
