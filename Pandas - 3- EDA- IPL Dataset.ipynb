```python
#find the most destructive death over batsman in the history of IPL
# find on the basis of strikerate, strikerate = (number of runs/number of balls)/100
# min, batsman should have played 200 balls in over 16-20
```

```python
df2.head()
```

```python
#find the dataframe which is deathover only. 
death_over = df2[df2['over'] > 15]
death_over
```

```python
all_batsman = df2.groupby('batsman')['batsman_runs'].count()    #count of balls batsman played in death over
all_batsman
```

```python
x = all_batsman  > 200                                          #filter who have played more than 200 balls
all_batsman[x]
```

```python
all_batsman[x].shape
```

```python
batsman_list = all_batsman[x].index.tolist()
batsman_list
```

```python
# calculate runs scored by all these batsman
#balls played by these batsman
```

```python
df2[df2['batsman'].isin(batsman_list)].shape
```

```python
final = df2[df2['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr
```
