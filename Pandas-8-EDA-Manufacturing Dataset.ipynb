### 2. Data Cleaning:

```python
df_cleaned = df.copy()
```

```python
#1. Drop fully null columns
df_cleaned.drop(columns=['CAMPAIGN_NBR'], inplace=True)
```

```python
# Fill missing categorical columns with mode
for col in df_cleaned.select_dtypes(include='object'):
    df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)
```

```python
# Fill missing numerical columns with median
for col in df_cleaned.select_dtypes(include=['int64', 'float64']):
    if df_cleaned[col].isnull().any():
        df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)
```

```python
column_analysis = pd.DataFrame({
    'Data Type': df_cleaned.dtypes,
    'Missing Values': df_cleaned.isnull().sum()
})
column_analysis
```

```python
### 3. Clean text values
categorical_cols = df_cleaned.select_dtypes(include='object').columns

for col in categorical_cols:
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip().str.lower()
```

```python
### 4. Remove outliers using IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
# Apply IQR filtering to key numerical columns
outlier_columns = ['KM', 'TOTALCOST', 'LBRCOST', 'REPORTING_COST']
for col in outlier_columns:
    if col in df_cleaned.columns:
        df_cleaned = remove_outliers_iqr(df_cleaned, col)
```

### 4. Generating tags/features from free text available:

```python
import pandas as pd
import re

# Load your cleaned data
df = df_cleaned

# Define common keyword patterns
failure_keywords = [
    'leak', 'noise', 'vibration', 'broken', 'faulty', 'issue', 'not working', 
    'fail', 'error', 'bubbling', 'peeling', 'fraying', 'off center', 'sticky', 
    'rubbing', 'clicking', 'cracked', 'coming apart', 'coming undone', 'delamination',
    'does not heat', 'lukewarm', 'not heating', 'inop', 'loose', 'not turning on'
]

component_keywords = [
    'engine', 'brake', 'mirror', 'door', 'window', 'sensor', 'seat', 'battery',
    'steering wheel', 'trim', 'bezel', 'stitching', 'leather', 'cover', 'spoke',
    'applique', 'button', 'switch', 'harness', 'heater', 'module','horn'
]

action_keywords = [
    'replaced', 'tightened', 'checked', 'cleaned', 'lubricated', 'repaired', 
    'realigned', 'adjusted', 'removed', 'reinstalled', 'inspected', 'reconnected',
    'tested', 'torqued', 'verified', 'submitted', 'retorqued', 'reassembled', 
    'returnrepalce'
]


# Preprocess text
def preprocess(text):
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    return text

df['CUSTOMER_VERBATIM_CLEAN'] = df['CUSTOMER_VERBATIM'].apply(preprocess)
df['CORRECTION_VERBATIM_CLEAN'] = df['CORRECTION_VERBATIM'].apply(preprocess)

# Function to generate tags
def generate_tags(text, keyword_list):
    return ', '.join([word for word in keyword_list if word in text])

# Apply tag generation
df['FAILURE_TAGS'] = df['CUSTOMER_VERBATIM_CLEAN'].apply(lambda x: generate_tags(x, failure_keywords))
df['COMPONENT_TAGS'] = df['CUSTOMER_VERBATIM_CLEAN'].apply(lambda x: generate_tags(x, component_keywords))
df['ACTION_TAGS'] = df['CORRECTION_VERBATIM_CLEAN'].apply(lambda x: generate_tags(x, action_keywords))
```

### Visuals from Tagged Columns

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
# Count tags directly from the FAILURE_TAGS column
failure_tags = df['FAILURE_TAGS'].dropna().str.cat(sep=', ').split(', ')
failure_tags = [tag.strip() for tag in failure_tags if tag.strip() != '']  # remove blanks
failure_counts = pd.Series(Counter(failure_tags)).sort_values(ascending=False)


plt.figure(figsize=(10, 5))
sns.barplot(x=failure_counts.values[:10], y=failure_counts.index[:10])
plt.title('Top 10 Failure Tags')
plt.xlabel('Count')
plt.ylabel('Failure Tag')
plt.tight_layout()
plt.show()
```

```python
# Count tags directly from the COMPONENT_TAGS column
component_tags = df['COMPONENT_TAGS'].dropna().str.cat(sep=', ').split(', ')
component_tags = [tag.strip() for tag in component_tags if tag.strip() != '']  # remove blanks
component_counts = pd.Series(Counter(component_tags)).sort_values(ascending=False)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=component_counts.values[:10], y=component_counts.index[:10])
plt.title('Top 10 Component Tags')
plt.xlabel('Count')
plt.ylabel('Component Tag')
plt.tight_layout()
plt.show()
```

```python
# Count tags directly from the ACTION_TAGS column
action_tags = df['ACTION_TAGS'].dropna().str.cat(sep=', ').split(', ')
action_tags = [tag.strip() for tag in action_tags if tag.strip() != '']  # remove blanks
action_counts = pd.Series(Counter(action_tags)).sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=action_counts.values[:10], y=action_counts.index[:10])
plt.title('Top 10 Action Tags')
plt.xlabel('Count')
plt.ylabel('Action Tag')
plt.tight_layout()
plt.show()
```
