```python
#SALE PRICE' column contains strings with ',' and '$'...remove them
df['SALE PRICE'] = df['SALE PRICE'].str.replace(',', '').str.replace('$', '').astype(float)
```

```python
# Drop the specified columns directly
df.drop(columns=['BOROUGH', 'EASE-MENT', 'APARTMENT NUMBER'], inplace=True)
```

```python
# Convert 'SALE DATE' to datetime
df['SALE DATE'] = pd.to_datetime(df['SALE DATE'], format='%d/%m/%Y')
```

```python
numerical_columns = ["BLOCK", "LOT", "ZIP CODE", "RESIDENTIAL UNITS", "COMMERCIAL UNITS", 
"TOTAL UNITS", "LAND SQUARE FEET", "GROSS SQUARE FEET", "YEAR BUILT","TAX CLASS AT TIME OF SALE", "SALE PRICE"]
df[numerical_columns] = df[numerical_columns].apply(pd.to_numeric, errors='coerce')
```

```python
 #Convert specified columns to categorical
categorical_columns = ['NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT', 
'BUILDING CLASS AT PRESENT', 'ADDRESS', 'BUILDING CLASS AT TIME OF SALE']

df[categorical_columns] = df[categorical_columns].astype('category')
```

```python
# Clean the categorical columns
for col in categorical_columns:
    df[col] = df[col].str.strip()  # Remove leading and trailing spaces
    df[col] = df[col].replace('', np.nan)  # Replace empty strings with NaN
```

```python
# Replace zeros with NaN in numerical columns
df[numerical_columns] = df[numerical_columns].replace(0, np.nan)
```

```python
# List of numerical columns
numerical_columns = ['ZIP CODE', 'TOTAL UNITS', 'SALE PRICE']

# Replace NaN values in numerical columns with the median of each column
for col in numerical_columns:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Taking the log of the "SALE PRICE" column
df['LOG_SALE_PRICE'] = np.log(df['SALE PRICE'])
```

```python
# Normalizing the numerical columns using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group the data by neighborhood and calculate the mean sale price for each neighborhood
neighborhood_prices = df.groupby('NEIGHBORHOOD')['SALE PRICE'].mean().reset_index()

# Sort the neighborhoods by mean sale price in descending order
neighborhood_prices = neighborhood_prices.sort_values(by='SALE PRICE', ascending=False)

# Create a bar plot to visualize the prices across neighborhoods
plt.figure(figsize=(16, 8))
sns.barplot(x='SALE PRICE', y='NEIGHBORHOOD', data=neighborhood_prices, palette='viridis')
plt.xlabel('Mean Sale Price')
plt.ylabel('Neighborhood')
plt.title('Mean Sale Price by Neighborhood')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

The purpose of this bar plot is to provide a visual comparison of the mean sale prices across different neighborhoods, allowing you to easily identify which neighborhoods tend to have higher or lower average sale prices based on the length of the bars. It helps in understanding the distribution of property sale prices within the dataset and identifying any significant variations between neighborhoods.

```python
# Extract month and year from the 'SALE DATE' column
df['SALE MONTH'] = df['SALE DATE'].dt.to_period('M').dt.strftime('%Y-%m')  # Convert to string format

# Group the data by month and calculate the mean sale price for each month
monthly_prices = df.groupby('SALE MONTH')['SALE PRICE'].mean().reset_index()

# Create a bar plot to visualize the mean sale prices by month
plt.figure(figsize=(12, 6))
plt.bar(monthly_prices['SALE MONTH'], monthly_prices['SALE PRICE'], color='skyblue')
plt.xlabel('Sale Month')
plt.ylabel('Mean Sale Price')
plt.title('Mean Sale Price by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns for the pairplot
numerical_columns = ['BLOCK', 'LOT', 'ZIP CODE', 'TOTAL UNITS', 'TAX CLASS AT TIME OF SALE', 'SALE PRICE']

# Subset the DataFrame to include only the selected numerical columns
df_numeric = df[numerical_columns]

# Create a pairplot
sns.pairplot(df_numeric, diag_kind='kde')  # diag_kind='kde' for kernel density plots on the diagonal
plt.show()
```

```python
# Select numerical columns for the correlation matrix
numerical_columns = ['BLOCK', 'LOT', 'ZIP CODE', 'TOTAL UNITS', 'TAX CLASS AT TIME OF SALE', 'SALE PRICE']

# Subset the DataFrame to include only the selected numerical columns
df_numeric = df[numerical_columns]

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
