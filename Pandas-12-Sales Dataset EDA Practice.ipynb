```python
#join orders and customers
#method1
pd.merge(orders, customers, on = 'OrderID', how = 'inner')
```

```python
#join orders and customers
#method2
orders.merge(customers, left_on = 'OrderID', right_on = 'OrderID', how = 'left')
```

```python
# stack columnwise both the orders and customers dataframe. use axis = 1 for columns and axis = 0 for row wise stack
pd.concat([orders,customers],axis = 1)
```

```python
# stack rowwise both the orders and customers dataframe. use axis = 1 for columns and axis = 0 for row wise stack
pd.concat([orders,customers],axis = 0)
```
